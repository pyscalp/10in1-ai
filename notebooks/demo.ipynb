# Nebius 10-in-1 AI Workbench — Quick smoke test

This notebook calls the Workbench Router endpoints.
Set `BASE_URL` and `TOKEN` first.

```python
import os, requests

BASE_URL = os.getenv("WORKBENCH_URL", "http://localhost:8000")
TOKEN = os.getenv("WORKBENCH_TOKEN", "change-me")
headers = {"Authorization": f"Bearer {TOKEN}", "Content-Type": "application/json"}
```

```python
# Health
requests.get(f"{BASE_URL}/health").json()
```

```python
# Crawl a page
requests.post(f"{BASE_URL}/crawl4ai/", headers=headers, json={"url": "https://example.com"}).json()
```

```python
# List LLM models (requires NEBIUS_LLM_URL to be configured)
requests.get(f"{BASE_URL}/llm/models", headers=headers).json()
```

```python
# Chat completion through the LLM endpoint
requests.post(
    f"{BASE_URL}/llm/chat/completions",
    headers=headers,
    json={
        "model": "Qwen/Qwen3-0.6B",
        "messages": [{"role": "user", "content": "Hello, who are you?"}]
    }
).json()
```
